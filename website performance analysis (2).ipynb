# Engaged vs non engaged sessions

```python
df.head()
```

```python
def create_engagement_comparison_chart(df: pd.DataFrame) -> None:
    """
    Creates and displays a bar chart comparing engaged vs non-engaged sessions by channel group.
    
    Args:
        df (pd.DataFrame): DataFrame containing session data with 'channel group', 
                          'Sessions', and 'Engaged sessions' columns.
    
    Returns:
        None: Displays the plot directly using plt.show()
    """
    # Calculate session metrics by channel group
    session_df = df.groupby("channel group")[["Sessions", "Engaged sessions"]].sum().reset_index()
    
    # Calculate non-engaged sessions
    session_df["Non-Engaged"] = session_df["Sessions"] - session_df["Engaged sessions"]
    
    # Prepare data for visualization by melting to long format
    session_df_melted = session_df.melt(
        id_vars="channel group",
        value_vars=["Engaged sessions", "Non-Engaged"]
    )
    
    # Create the visualization
    plt.figure(figsize=(8, 5))
    sns.barplot(
        data=session_df_melted,
        x="channel group",
        y="value",
        hue="variable"
    )
    
    # Add chart formatting
    plt.title("Engaged vs Non-Engaged Sessions by Channel Group")
    plt.xticks(rotation=45)
    plt.ylabel("Number of Sessions")
    plt.tight_layout()  # Ensures labels fit within the figure
    
    # Display the chart
    plt.show()

# Usage example:
# create_engagement_comparison_chart(df)
```
